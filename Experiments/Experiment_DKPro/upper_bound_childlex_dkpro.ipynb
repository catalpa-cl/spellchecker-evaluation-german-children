# Upper bound of childlex-litkey which was used by DKPro

Upper bound for boyd-analysis is created by comparing target words ('corrected') with dictionary that is given by the user to boyd (in this case it is childlex).


Procedure:
- Dictionary: Childlex
- Target words (corrected types & token)
- Is target in dict?

Outline:
- Preparation
- Get upper bound
- Difference childlex - target

## Get upper bound
### Case sensitive

```python
# Types
dict_count_types = types.loc[types.corrected.isin(childl.word)]
print("Count type: ",types.shape[0])
print("Count childl in types :",dict_count_types.shape[0])
```

```python
# Token
dict_count_token = token.loc[token.corrected.isin(childl.word)]
print("Count token: ",token.shape[0])
print("Count childl in token:",dict_count_token.shape[0])
```

```python
#display(childl.loc[childl.word.isin(token.corrected)])
#display(childl.shape[0])
```

```python
# Ratio types
ratio_types = dict_count_types.shape[0]/types.shape[0]
print('Upper bound for types is', round(ratio_types*100,4),'%')
print(ratio_types)
```

```python
# Ratio token
ratio_token = dict_count_token.shape[0]/token.shape[0]
print('Upper bound for token is', round(ratio_token*100,4),'%')
print(ratio_token)
```

### Case insensitive

```python
token_lc = token.corrected.str.lower()
types_lc = types.corrected.str.lower()

childl_lc = childl.word.str.lower()
```

```python
childl_lc.head(20)
```

```python
# Types lowercase
dict_count_types_lc = types_lc.loc[types_lc.isin(childl_lc)]
#display(dict_count_types_lc.shape[0])

print("Count type lc: ",types_lc.shape[0])
print("Count childl in types lc:",dict_count_types_lc.shape[0])
```

```python
# Token lowercase
dict_count_token_lc = token_lc.loc[token_lc.isin(childl_lc)]
#display(dict_count_types_lc.shape[0])

print("Count token lc: ",token_lc.shape[0])
print("Count childl in token lc:",dict_count_token_lc.shape[0])
```

```python
# Ratio types lowercase
ratio_types_lc = dict_count_types_lc.shape[0]/types_lc.shape[0]
print('Upper bound for types LC is', round(ratio_types_lc*100,4),'%')
print(ratio_types_lc)
```

```python
# Ratio token lowercase
ratio_token_lc = dict_count_token_lc.shape[0]/token_lc.shape[0]
print('Upper bound for token LC is', round(ratio_token_lc*100,4),'%')
print(ratio_token_lc)
```
