### Further exploring

```python
# Childlex
childlex = r'../../Lexicons/childlex_0.17.01_all_types.txt'

# Read in and change dtype to string, otherwise object (thus could not compare strings)
with open(childlex, mode="r", encoding="utf-8") as f:
    chl = f.read().splitlines()
    chl = set(chl)
    
chl_lower = set([])
for c in chl:
    chl_lower.add(c.lower())

```

```python
origs = list(data_error_types_not_inSugg["original"].values)
targets = list(data_error_types_not_inSugg["corrected"].values)
pairs = zip(origs, targets)
```

Number of types which were not in suggestions

```python
len(targets)
```

How many target types (lowercased) are not in childlex?

```python
count_not_in_childlex = sum(1 for t in targets if t.lower() not in chl_lower)
count_not_in_childlex
```

```python
round(count_not_in_childlex/len(targets),3)
```

How many target types (lowercased) are in childlex and thus could have been corrected?

```python
len(targets)-count_not_in_childlex
```

Exploring those which are in childlex

```python
in_childlex = []
for pair in zip(origs,targets):
    if pair[1].lower() in chl_lower:
        in_childlex.append(pair)
```

Proportion of not corrected words although the target is in childlex

```python
len(in_childlex)/len(data_error_types)

```

```python
for pair in in_childlex:
    print(pair)
```
